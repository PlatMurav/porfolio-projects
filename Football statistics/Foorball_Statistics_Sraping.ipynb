# More than one team

Now we need just to create a loop and collect information about all teams in the table

```python
years = list(range(2023, 2020, -1)) # all years we want to analyse
```

```python
# It will contain several DataFrames and each DataFrame is going to contain the match log for one team in one season

all_matches = [] 
```

```python
standings_url = 'https://fbref.com/en/comps/9/Premier-League-Stats'
```

##### All code in the next section just repeats all steps we did before and iterates tgrough every year

```python
for year in years:
    data = requests.get(standings_url)
    soup = BeautifulSoup(data.text)
    standings_table = soup.select('table.stats_table')[0]
    
    links = [l.get("href") for l in standings_table.find_all('a')]
    links =  [l for l in links if '/squads/' in l]
    team_urls = [f"https://fbref.com{l}" for l in links] # we get all links, related to diffrent teams
    
    previous_season = soup.select('a.prev')[0].get('href')
    standings_url = f"https://fbref.com{previous_season}"
    
    for team_url in team_urls:
        team_name = team_url.split('/')[-1].replace('-Stats', '').replace('-','')
        
        data = requests.get(team_url)
        matches = pd.read_html(data.text, match = 'Scores & Fixtures')[0]
        
        soup = BeautifulSoup(data.text)
        links = [l.get('href') for l in soup.find_all('a')]
        links = [l for l in links if l and 'all_comps/shooting/' in l]
        data = requests.get(f'https://fbref.com{links[0]}')
        shooting = pd.read_html(data.text, match = 'Shooting')[0]
        shooting.columns = shooting.columns.droplevel()
        
        try:
            team_data = matches.merge(shooting[['Date', 'Sh', 'SoT', 'Dist', 'FK', 'PK', 'PKatt']], on = 'Date')
        except ValueError:
            continue
            
        team_data = team_data[team_data['Comp'] == "Premier League"]
        team_data['Season'] = year # we need to know what year was it
        team_data['Team'] = team_name
        all_matches.append(team_data)
        
        time.sleep(3) # in order not to make a lot of requests very quickly and slow down websites, I don't wanna be banned
        
```

```python
match_df = pd.concat(all_matches) # we concatenate all tables in the list
```

### Second Model and Rolling averages

We can to compute **rolling averages** for team perfomance: how this team did in the past few games, how many shots they had, how many goals were scored against them

It is done to to discover the trend, a rolling average uses smaller parts of the data. For instance, you might use the numbers collected over the last 30 days and find out their average. Then, the average rolls or moves for each new period. Making these calculations for every 30 day period allows professionals to discover the way the average changes as time moves forward.

```python
grouped_matches = matches.groupby('team') # one DataFrame for every squad in our data
```

```python
type(grouped_matches)
```

It will help us to analyse their statistic (**for exmaple**, if a team lost 4 matches in row, it is highly likely that it will lost again because all players might not be in a good shape)

```python
group = grouped_matches.get_group("Arsenal") 
```

```python
group.head()
```

We should create a function that takes a group, a set of columns that we want to compute rolling averages for and then it'll take a new set of columns that we want to assign rolling averages to

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values('date')
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats # extra columns that describe what happened in the previous 3 matches
    group = group.dropna(subset = new_cols) # we drop any missing values (any rows that have missing values)
    return group
```

```python
cols = ['gf', 'ga', 'sh', 'sot', 'dist', 'fk', 'pk', 'pkatt']
new_cols = [f"{c}_rolling" for c in cols]
```

gf - goals for

ga - goals agains

sh - shots

sot - shots on target

dist - total distance each shot travelled

fk - free kicks

pk - penalty kicks

pka - penalty kick attempts

```python
rolling_averages(group, cols, new_cols).head() # we added the columns that describe what happened in the previous 3 matches
```

```python
# We compute rolling averages for every team

matches_rolling = grouped_matches.apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
grouped_matches.apply(lambda x: rolling_averages(x, cols, new_cols)).head()
```

```python
matches_rolling = matches_rolling.droplevel('team') # we drop extra index level (team)
```

```python
matches_rolling
```

As we can see we have 1705 rows but our index reaches only 14 at the end, it means a lot of values in our index are repeated

```python
matches_rolling.index = range(matches_rolling.shape[0]) # we want unique values for each index
```

```python
train = matches_rolling[matches_rolling['date'] < "2023-01-01"]
test = matches_rolling[matches_rolling['date'] > "2023-01-01"]
```

```python
rfc.fit(train[predictors + new_cols], train["target"])
```

```python
preds = rfc.predict(test[predictors + new_cols])
```

```python
print("Accuracy:", accuracy_score(test['target'], preds))
print("Precision:", precision_score(test['target'], preds))
print("Recall:", recall_score(test['target'], preds))
```
